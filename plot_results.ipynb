```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("results.csv")
```

# Scatter plot of time taken vs number of reads: 
This plot will show how the time taken by each method scales with the number of reads in the input BAM file.

```python
sns.scatterplot(data=df, x="num_reads", y="time", hue="experiment_name")
plt.xlabel("Number of Reads")
plt.ylabel("Time (s)")
plt.title("Time Taken vs Number of Reads")
plt.legend(title="Method", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Bar plot of average time taken by each method: 
This plot will show the average time taken by each method across all input BAM files.

```python
avg_time = df.groupby("experiment_name")["time"].mean()

sns.barplot(x=avg_time.index, y=avg_time.values)
plt.xlabel("Method")
plt.ylabel("Average Time (s)")
plt.title("Average Time Taken by Each Method")
plt.xticks(rotation=45, ha="right")
plt.show()
```

# Scatter plot of memory usage vs number of reads: 
This plot will show how the memory usage of each method scales with the number of reads in the input BAM file.

```python
sns.scatterplot(data=df, x="num_reads", y="max_mem", hue="experiment_name")
plt.xlabel("Number of Reads")
plt.ylabel("Memory Usage (KB)")
plt.title("Memory Usage vs Number of Reads")
plt.legend(title="Method", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Bar plot of average memory usage by each method: 
This plot will show the average memory usage of each method across all input BAM files.

```python
avg_mem = df.groupby("experiment_name")["max_mem"].mean()

sns.barplot(x=avg_mem.index, y=avg_mem.values)
plt.xlabel("Method")
plt.ylabel("Average Memory Usage (KB)")
plt.title("Average Memory Usage by Each Method")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Sort the dataframe by time taken in ascending order
df_sorted = df.sort_values('time')

# Select the top 3 and bottom 3 methods
top_3 = df_sorted.head(3)['experiment_name'].unique()
bottom_3 = df_sorted.tail(3)['experiment_name'].unique()

# Subset the data for the top 3 and bottom 3 methods
df_top_3 = df[df['experiment_name'].isin(top_3)]
df_bottom_3 = df[df['experiment_name'].isin(bottom_3)]
```

# Top 3 Fastest Wall Time

```python
sns.scatterplot(data=df_top_3, x="num_reads", y="time", hue="experiment_name")
plt.xlabel("Number of Reads")
plt.ylabel("Time (s)")
plt.title("Time Taken vs Number of Reads (Top 3 Methods)")
plt.legend(title="Method", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Bottom 3 wall time

```python
sns.scatterplot(data=df_bottom_3, x="num_reads", y="time", hue="experiment_name")
plt.xlabel("Number of Reads")
plt.ylabel("Time (s)")
plt.title("Time Taken vs Number of Reads (Bottom 3 Methods)")
plt.legend(title="Method", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
